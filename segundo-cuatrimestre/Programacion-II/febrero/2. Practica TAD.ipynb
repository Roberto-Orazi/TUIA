_Nota: no olvidar describir el \_\_init___

````python
class celda:
    def __init__(self,valor):
        self.valor=valor
    def __str__(self):
        return(f'saludos puto {self.valor}')
    def nuevo_valor(self,valor_nuevo):
        self.valor=valor_nuevo
        return str(self.valor)

mati=celda('asdasd')
print(mati)
mati.nuevo_valor('chaaa')
print(mati)
````
